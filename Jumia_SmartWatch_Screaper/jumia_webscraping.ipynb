```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime   # for time stamp
import time                     # to make it sleep
```

```python
link = 'https://www.jumia.com.eg/catalog/?q=smart+watch'
product_details = []
```

```python
def jumia_scraper():
    
    page = requests.get(link)
    soup = BeautifulSoup(page.text, 'html.parser')
    product_list = soup.find_all('article', {'class':'prd'})
    print(product_list)
    for product in range(len(product_list)):
        try:
            prod_name = product_list[product].find('h3', {'class':'name'}).text
        except:
            prod_name = 'NotFound'
        try:
            prod_price = product_list[product].find('div', {'class':'prc'}).text
        except:
            prod_price = 'NotFound'
        try:
            prod_rate= product_list[product].find('div', {'class':'stars'}).text
        except:
            prod_rate = 'NotFound'
        try:
            prod_discount = product_list[product].find('div', {'class':'bdg'}).text
        except:
            prod_discount = 'NotFound'
    
        prod_link = f"https://www.jumia.com.eg{product_list[product].find('a', {'class':'core'}).get('href')}"

    # time stamp for scraping product
    MonthStamp = datetime.now().strftime('%m')
    DayStamp = datetime.now().strftime('%d')
    HourStamp = datetime.now().strftime('%H')
    product_details.append({
        'website': 'Jumia',
        'prod_name': prod_name,
        'prod_price': prod_price,
        'prod_rate' : prod_rate,
        'prod_discount': prod_discount,
        'prod_link' : prod_link,
        'scrape_MonthStamp': MonthStamp,
        'scrape_DayStamp': DayStamp,
        'scrape_HourStamp': HourStamp
    })
```

```python
def Saving_toCSV()
    
    data = pd.DataFrame(product_details)
    data.head(20)
    data.to_csv("Jumia_smart_watch.csv", header=True)
    print(" Data Loaded To CSV Successfully")
```

```python
import schedule
import time

def daily_task():
    jumia_scraper()
    Saving_toCSV()

# Schedule tasks (Monday-Friday at 9:00 AM)
schedule.every().monday.at("09:00").do(daily_task)
schedule.every().tuesday.at("09:00").do(daily_task)
schedule.every().wednesday.at("09:00").do(daily_task)
schedule.every().thursday.at("09:00").do(daily_task)
schedule.every().friday.at("09:00").do(daily_task)

while True:
    schedule.run_pending()
    time.sleep(60)  # Check every minute

# nohup python scheduler.py &  # Linux/Mac
```
