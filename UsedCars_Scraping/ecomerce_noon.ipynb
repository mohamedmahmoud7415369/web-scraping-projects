```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time
from datetime import datetime
import pandas as pd
import seaborn
import matplotlib
import csv
import json
```

```python
#input('what is the category of products you want scrapping? : ')
Categories = ['Electronics', 'Mobiles', 'Tools', 'Books']
pages_number =  10
product_details = []
```

```python
def noon():
    try:
        service = Service(ChromeDriverManager().install())
        browser = webdriver.Chrome(service=service)
        for category in Categories:
            for n in range(1, pages_number):
                url = f'https://www.noon.com/egypt-en/search/?q={category}&page={n}'
                browser.get(url)
                browser.maximize_window()
                time.sleep(3)
                #search_bar = browser.find_element('id', 'search-input').send_keys('smart watch')
                #time.sleep(3)

                product_list = browser.find_elements('class name', 'PBoxLinkHandler_linkWrapper__8FlHB')
                print(product_list)

                for product in product_list:
                    html_code = product.get_attribute('outerHTML')
                    soup = BeautifulSoup(html_code, 'html.parser')
                    try:
                        Prod_name = soup.find('h2', {'class':'ProductDetailsSection_title__JorAV'}).text
                    except:
                        Prod_name = 'Not Found'
                    try:
                        Current_price = soup.find('strong', {'class':'Price_amount__2sXa7'}).text
                    except:
                        Current_price = 'Not Found'
                    try:
                        Review = soup.find('div', {'class':'RatingPreviewStar_textCtr__sfsJG'}).text
                    except:
                        Review = 'No review'
                    try:
                        Num_comments = soup.find('div', {'class':'RatingPreviewStar_countCtr__p_2z7'}).text
                    except:
                        Num_comments = 'No comments'
                    try:
                        currency = soup.find('span', {'class':'Price_currency__taKQj'}).text
                    except:
                        currency = 'Not Found'
                    try:
                        Old_price = soup.find('span', {'class':'Price_oldPrice__ZqD8B'}).text
                    except:
                        Old_price = 'Not Found'
                    try:
                        discound = soup.find('span', {'class':'PriceDiscount_discount__1ViHb PriceDiscount_pBox__eWMKb'}).text
                    except:
                        discound = 'No discound'
                    try:
                        delivery = soup.find('div', {'class':'Nudges_nudgesContainer__vC72M'}).text
                    except:
                        delivery = 'Not Found'
                    try:
                        Get_date = soup.find('div', {'class':'ProductDetailsSection_deliveryDate__8BEgY'}).text
                    except:
                        Get_date = 'Not Found'
                    try:
                        fulfiment = soup.find('div', {'class':'FulfilmentBadge_container__Fzk3K FulfilmentBadge_small__eUX03'}).text
                    except:
                        fulfiment = 'Not Found'
                    try:
                        best_seller = soup.find('div', {'class':'BestSellerTag_container__q2YWE'}).text
                    except:
                        best_seller = 'Not best seller'
                    try:
                        Prod_link = soup.find('a').get('href')
                    except:
                        Prod_link = 'No link Found'
                    
                    product_details.append({
                        'Category':category,
                        'Prod_name':Prod_name,
                        'Current_price':Current_price,
                        'Num_comments':Num_comments,
                        'Review':Review,
                        'Old_price':Old_price,
                        'discound':discound,
                        'currency':currency,
                        'delivery':delivery,
                        'Get_date':Get_date,
                        'fulfiment':fulfiment,
                        'best_seller':best_seller,
                        'Prod_link':Prod_link
                    })
                
                print(len(product_details))

    except Exception as e:
        print(f'something went wrong with Noon ==> {e}')
    browser.close()
```

```python
def Printing_file():
    try:
        path = './Noon_product.csv'
        column = product_details[0].keys()
        with open(path, 'w', newline='', encoding='UTF-8') as output_file:
            dict_writer = csv.DictWriter(output_file, column)
            dict_writer.writeheader()
            dict_writer.writerows(product_details)
        print('File Printed Successfully')
    except Exception as e:
        print(f'something went wrong with Printing_file ==> {e}')
```
