```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time
from datetime import datetime
import pandas as pd
import seaborn
import matplotlib
import csv
import json
```

```python
#input('what is the category of products you want scrapping? : ')
pages_number = 199
car_details = []
```

```python
def Dubizzel():
    try:
        service = Service(ChromeDriverManager().install())
        browser = webdriver.Chrome(service=service)
        for n in range(1, pages_number):
            url = f'https://www.dubizzle.com.eg/en/vehicles/cars-for-sale/used/?page={pages_number}&filter=price_between_0_to_5000000%2Cyear_between_2000_to_2026'
            browser.get(url)
            browser.maximize_window()
            time.sleep(3)
            #search_bar = browser.find_element('id', 'search-input').send_keys('smart watch')
            #time.sleep(3)

            car_list = browser.find_elements('class name', '_70cdfb32')
            print(car_list)

            for car in car_list:
                    html_code = car.get_attribute('outerHTML')
                    soup = BeautifulSoup(html_code, 'html.parser')
                    try:
                        Car_Price = soup.find('span', {'class':'ddc1b288'}).text
                    except:
                        Car_Price = 'Not Found'
                    try:
                        Car_name = soup.find('h2', {'class':'_562a2db2'}).text
                    except:
                        Car_name = 'Not Found'
                    try:
                        Car_year = soup.find('span', {'class':'_18b01e88'}).text
                    except:
                        Car_year = 'No Car_year'
                    try:
                        Mileage = soup.find('span', {'class':'RatingPCar_yearStar_countCtr__p_2z7'}).text
                    except:
                        Mileage = 'No comments'
                    try:
                        FuelType = soup.find('span', {'class':'_3e1113f0'}).text
                    except:
                        FuelType = 'Not Found'
                    try:
                        Location = soup.find('span', {'class':'f7d5e47e'}).text
                    except:
                        Location = 'Not Found'
                    try:
                        Creation_date = soup.find('span', {'class':'c72cec28 undefined'}).text
                    except:
                        Creation_date = 'No Creation date'
                    try:
                        Contect_method = soup.find('span', {'class':'_30de236c af478541 b7af14b4'}).text
                    except:
                        Contect_method = 'Not Found'
                    
                    car_details.append({
                        'Car_Price':Car_Price,
                        'Car_name':Car_name,
                        'Mileage':Mileage,
                        'Car_year':Car_year,
                        'Location':Location,
                        'Creation_date':Creation_date,
                        'FuelType':FuelType,
                        'Contect_method':Contect_method
                    })
                
            print(len(car_details))

    except Exception as e:
        print(f'something went wrong with Dubizzel ==> {e}')
    browser.close()
```

```python
def Printing_file():
    try:
        path = './Dubizzle_product.csv'
        columns = car_details[0].keys()
        with open(path, 'w', newline='', encoding='UTF-8') as output_file:
            dict_writer = csv.DictWriter(output_file, columns)
            dict_writer.writeheader()
            dict_writer.writerows(car_details)
        print('File Printed Successfully')
    except Exception as e:
        print(f'something went wrong with Printing_file ==> {e}')
```

```python
Dubizzel()
Printing_file()
```
